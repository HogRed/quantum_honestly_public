## Same in, different out

Ten electrons. **Same** magnet, **same** dial, **same** everything — fired one at a time.

The reading stutters: **up, down, down, up, down, up, down…**

Identical preparation, different answers — *every time.* Either I'm fumbling the experiment, or the universe is **rolling dice.** Today we find the dice, and the one rule that says how they're loaded.

Cold open: the puzzle of same-in / different-out. No definitions yet — just the stutter of outcomes.

## Where we left off — the quantum coin

From **E2**, fresh in mind before we measure:

- A qubit = **two arrows**, `[a, b]` — `a` for **up**, `b` for **down** — held **both at once** (*superposition*) until you look.
- Square each arrow's **length** → the **odds**: `|a|²` and `|b|²`, and they total **100%** (`|a|² + |b|² = 1`).
- On the **Bloch globe** the state is an arrow on a sphere; its **phase** is just how that arrow is spun.
- E2's coin: a *secret-answer* coin matches the odds on the up/down (**Z**) axis — but turn the magnet to left/right (**X**) and it **fails.** The already-decided coin *looks* dead.

> **Leak:** *looks* dead, not *proven* dead — that airtight proof is **Bell's theorem, owed in E5.** And the one promise E2 left open — *why* square the length? — we pay **today.**

## The trap  ⚠ *the coin trap (2 of 3)*

The comfortable story: *'the answer was decided all along — the electron was secretly already up or down, and the probability is just **us not knowing yet.**'* A coin already lying under a cup.

That's ordinary **ignorance** — a fact that exists, you just haven't peeked. For a quantum coin it is **wrong**: it genuinely has *no face* until you look.

> **Leak:** the coin-under-a-cup is a perfect picture of *classical* probability — right up to the punchline. The airtight proof it can't secretly have a face is **Bell's theorem — owed in E5.**

## The Born rule: square the arrow's length

<video src="media/videos/E3_manim/1080p60/BornShadow.mp4" width="820" controls autoplay loop muted></video>

`P(up) = |a|²`,  `P(down) = |b|²`. The amplitude is a little arrow; the probability is its **shadow** — the squared length. `|a|² = a·a*` — multiply the arrow by its mirror image and the angle cancels, leaving a real, non-negative number. Exactly what an odds has to be.

> **Leak:** you can see a real arrow whole; you only ever get the quantum arrow's *shadow* — one number, the squared length — and only statistically. The shadow throws the **phase** away.

## 🎓 A closer look: why *squared*?

The two odds must total 100% no matter which axis you turn the magnet to, and the squared length is the ONLY thing that stays pinned when you rotate. Notice `|a|² + |b|² = 1` is just Pythagoras — the equation of a circle of radius 1: turn your frame and `a` and `b` change, but `a²+b²` doesn't budge. If nature had used `|a|` or `|a|³`, the total would drift off 100% the instant you rotated, and probability would leak out of the universe.

**Picture it —** GPS distance: the east-leg and north-leg of a trip change if you rotate your map, but the straight-line distance — square the legs, add, square-root — never changes. The Born rule is “Pythagoras for probability.”

**Watch out:** squaring isn't a convention physicists picked — it's the unique measure that keeps total probability at 100% under every rotation (Gleason's theorem is the formal version).

<small>Source: Tong §1.1 p.4 (P=|ψ|², 'after Max Born who first understood that quantum mechanics is a theory of probability rather than certainty') and §3.3 eq 3.10 p.76; Sean Carroll, 'Why Probability in QM is Given by the Wave Function Squared' https://www.preposterousuniverse.com/blog/2014/07/24/why-probability-in-quantum-mechanics-is-given-by-the-wave-function-squared/ ; 'Born is the offspring of Pythagoras' arXiv:1905.08429 https://arxiv.org/pdf/1905.08429 ; Quanta Magazine on the Gleason/Born reconstruction https://www.quantamagazine.org/the-born-rule-has-been-derived-from-simple-physical-principles-20190213/</small>

## Outcomes, axes & collapse

<video src="media/videos/E3_manim/1080p60/CollapseSnap.mp4" width="820" controls autoplay loop muted></video>

Take the upright **Z detector** from E2 and write it as numbers: a little **2×2 grid** — a question the math can ask. We call that grid an **observable**, and the two numbers down its diagonal, **+1** and **−1**, *are* the two readings.

`Z = [[1,0],[0,-1]]` carries a menu: feed it *up* → **+1**·up, *down* → **−1**·down. Those readings (**+1, −1**) are the **eigenvalues** — the outcomes; up & down are the **eigenvectors** — the axes. The instant the dial reads, the state **snaps** to that axis and parks there — **that snap is collapse.** Ask again, same answer.

> **Leak (snap-to-grid):** a drawing app snaps to the *nearest* dot — deterministic. Collapse is **random**, weighted by `|amplitude|²`, and it **erases** where you were. The cursor remembers; the qubit forgets. *(And 'measure' needs a **detector**, not a **mind** — that's decoherence.)*

## 🎓 A closer look: observable, eigenvalue, eigenvector

An *observable* (here `Z = [[1,0],[0,-1]]`) is the math stand-in for the question “are you up or down?” — a grid of numbers that *acts on* a two-slot state.

**How a grid eats a list:** the top row `[1, 0]` reads the up-slot and keeps it; the bottom row `[0, −1]` reads the down-slot and flips its sign. So pure up `(1,0)` comes back as `(+1)(1,0)` — itself, times **+1**; pure down `(0,1)` comes back as `(−1)(0,1)` — itself, times **−1**.

Feed it a *tilted* state and it scrambles it — but those two special states come back unchanged except for a multiplier. The un-scrambled survivors are the **eigenvectors** (the axes), and the multipliers `+1, −1` are the **eigenvalues** (the only numbers the dial can ever show). Because observables are *Hermitian* — a **balanced grid** whose mirror-flip equals itself — those eigenvalues are guaranteed **real**: plain dots on a line, the kind a dial can show, not the arrows from E2.

**Picture it —** a vending machine with a fixed button panel: you can press anywhere on the glass, but only the labeled buttons (the menu = eigenvalues) dispense anything, each giving one fixed item.

**Watch out:** a tilted state can't make the dial read something “in between” like −0.3. The only allowed readings are the eigenvalues (+1 or −1); the tilt only sets the ODDS of each.

<small>Source: Tong §3.2.1 p.68 (boxed law 'The outcome of any measurement of Ô lies in the spectrum of Ô'; eigenvalue equation Ôψ=λψ; the word 'spectrum') and §3.2.2 p.71-72 (Hermitian operators; 'The eigenvalues λn are real'); §3.3 eq 3.10 p.76.</small>

## 🎓 A closer look: a detector, not a mind

Pop-science says a conscious observer freezes the quantum world by looking. That's not physics. Collapse is triggered by ANY irreversible interaction with a large environment — a detector firing, a CCD pixel lighting up, a single air molecule or photon bumping the electron. The name for that environmental scrambling is **decoherence**. Be honest about the gap: decoherence explains why we never SEE superpositions of big things, but not exactly WHEN/WHERE one definite outcome gets singled out — that leftover is the genuine, still-unsolved “measurement problem.”

**Picture it —** a secret whispered in a crowded room: while only you know it, it can be taken back; the instant it's repeated to a hundred people it's irreversibly out — copied into the environment, no mind required.

**Watch out:** consciousness does NOT cause collapse — any irreversible interaction with the environment does. Eugene Wigner himself dropped the consciousness idea after H. D. Zeh's 1970 decoherence work.

<small>Source: Tong §1.2.3 p.13 (collapse; a detector placed near X₁) and §3.3 p.78 ('measurement of a quantum system is never innocent. You can't just take a small peek and walk away pretending that you've not done anything') ; Tong §3.5.2 p.92 (quantum decoherence paragraph; 'falls short of resolving all the difficulties of either interpretation') ; Wikipedia, 'Consciousness causes collapse' (Wigner/Zeh history) https://en.wikipedia.org/wiki/Consciousness_causes_collapse</small>

## Phase is invisible to one shot

<video src="media/videos/E3_manim/1080p60/PhaseInvisible.mp4" width="820" controls autoplay loop muted></video>

Spin the down-amplitude's phase a full turn. `|b|²` — the odds bar — **doesn't budge.** Squaring washes the phase out.

> **But don't throw it away:** the universe keeps the phase. It's the whole two-slit pattern from E1 — invisible to one measurement, decisive across two.

## 🎓 A closer look: amplitude is not the probability

Two words that sound like synonyms but aren't: the **amplitude** (the spinning arrow `a`) versus the **probability** (the odds you get only AFTER squaring it). The amplitude carries two facts at once — a length and a direction (the phase). Squaring keeps the length and discards the direction, which is exactly why one measurement can never reveal the phase. But “invisible to one shot” is not “deleted”: the universe keeps the whole arrow, and the phase comes roaring back the moment two arrows are ADDED before squaring — that's interference.

**Picture it —** a bank balance: a $5 debt and a $5 credit have the same size but opposite sign. Record only the magnitude and you can't tell them apart — yet that sign is exactly what lets two amounts cancel to zero. That cancellation is interference.

**Watch out:** the phase isn't “deleted because squaring ignores it.” It's invisible to ONE measurement but decisive across two — it's what makes interference, so it's fully real.

<small>Source: Tong §1.1 p.3-4 (the state is a complex-valued wavefunction; the mod-square is the probability) ; course glossary distinction 'probability vs probability amplitude — odds vs the complex arrow you square to GET the odds; keep them straight' (course-outline.md, E3 'New jargon glossed'); reinforces the E1/E2 mantra 'amplitudes add, THEN we square.'</small>

## The law, boxed

$$P(\text{up}) = |a|^2 \qquad P(\text{down}) = |b|^2$$
$$|a|^2 + |b|^2 = 1 \quad\text{(odds total 100%)}$$
$$\langle Z \rangle = (+1)|a|^2 + (-1)|b|^2 \quad\text{(average reading)}$$

Those angle brackets `⟨ ⟩` mean **“the average of”** — a different job than E2's overlap brackets `⟨φ|ψ⟩`.

Three lines and you have measurement. *Borrowed, flagged honestly:* **why squared** is an axiom (name-drop **Gleason**); real readings come from **Hermitian** matrices (a *balanced grid* — its mirror-flip equals itself — which forces real outcomes); *inherent ≠ ignorance* is **Bell, owed in E5**; and `P(x)=|ψ(x)|²` scales the same rule to the wavefunction (**E6**).

## 🎓 A closer look: ⟨Z⟩ is an average, not a reading

`⟨Z⟩ = (+1)(0.25) + (−1)(0.75) = −0.5` is the probability-weighted average of the two readings — the “average needle” you'd get by firing thousands of identical electrons and averaging. Here's the trap: the dial can ONLY ever read +1 or −1, so it will never actually show −0.5 on any single shot. The expectation value is a summary of a whole crowd of measurements, not a possible outcome of one — exactly like a fair die averaging 3.5 though no face shows 3.5.

**Picture it —** “the average household has 1.8 children.” No actual household contains 1.8 kids — it's a property of the whole population, never of any single member.

**Watch out:** ⟨Z⟩ = −0.5 does NOT mean a shot can read −0.5, or that the electron is “half-down.” A single measurement yields only an eigenvalue (+1 or −1); −0.5 is just their long-run weighted average.

<small>Source: Tong §3.3.1 p.79 (expectation value ⟨Ô⟩ψ = Σ|aₙ|²λₙ, eq 3.13; 'if we have many systems, all in the same state ψ, and perform the same measurement on each then the results will differ but the average will be…') ; Wikipedia, 'Expectation value (quantum mechanics)' (the expectation value 'may have zero probability of occurring' — e.g. integer-only outcomes with a non-integer mean) https://en.wikipedia.org/wiki/Expectation_value_(quantum_mechanics) ; student-difficulties study arXiv:1701.01409 https://arxiv.org/pdf/1701.01409</small>

## Let's actually do it — by hand

State `|ψ⟩ = (a, b) = (1/2, √3/2)`, with `√3/2 ≈ 0.866`.

- `P(up)   = |1/2|²  = 0.250` → **25%**
- `P(down) = |√3/2|² = 0.750` → **75%**
- check: `0.250 + 0.750 = 1.000` ✓   ·   out of 1000: ~**250 up, 750 down**
- `⟨Z⟩ = (+1)(0.250) + (-1)(0.750) = -0.500`  (the average needle)

Now let the toolkit confirm it — same `qm.born_probs`, `qm.expectation`.

```python
# LIVE: square each amplitude -> the odds (Born rule), and the phase twist
for name, st in [("psi  = (1/2, sqrt3/2) ", psi), ("psi' = (1/2, i*sqrt3/2)", psi_phase)]:
    pz = qm.born_probs(st, "Z")
    print(f"{name}:  P(up) = {pz[+1]:.3f}   P(down) = {pz[-1]:.3f}   <Z> = {qm.expectation(st, 'Z'):+.3f}")
print("\nphase is invisible to one measurement: the odds are IDENTICAL (squaring kills the i).")
```

## Roll it 10,000 times

The squared lengths *are* the long-run odds. Fire `psi` through the Z-detector 10,000 times with `qm.born_sample`; watch the running fraction crawl to **0.25** and the average needle to **-0.5**.

```python
# LIVE: roll the quantum die 10,000 times -- the wobble locks onto 0.25
qm.seed(2)
shots = 10000
outs  = qm.born_sample(psi, "Z", shots=shots)        # +1 = up, -1 = down
run_up = np.cumsum(outs == +1) / np.arange(1, shots + 1)
run_z  = np.cumsum(outs)       / np.arange(1, shots + 1)
print(f"measured fraction up = {np.mean(outs == +1):.4f}    (Born: |a|^2 = 0.25)")
print(f"sample average <Z>   = {outs.mean():+.4f}   (Born: <Z> = -0.5)")
fig, (axu, axz) = plt.subplots(1, 2, figsize=(10, 3.6))
axu.plot(np.arange(1, shots + 1), run_up, color="#2b8cbe", lw=1); axu.axhline(0.25, ls="--", c="k")
axu.set_xscale("log"); axu.set_ylim(0, 1); axu.set_title("running fraction up -> 0.25")
axz.plot(np.arange(1, shots + 1), run_z, color="#859900", lw=1); axz.axhline(-0.5, ls="--", c="k")
axz.set_xscale("log"); axz.set_ylim(-1, 1); axz.set_title("running average <Z> -> -0.5")
plt.tight_layout(); plt.show()
```

## Collapse, made visible

Fire one shot. It reads **down** → the state snaps to `(0, 1)`. Re-measure: `P(down) = |1|² = 1` — dead certain. And the tilted state is a **pin on the globe** at 120° off north *(cos 60° = ½, and from E2 the globe doubles the angle — so 60° becomes 120°)*; measuring snaps it to a **pole**, never in between.

```python
# LIVE: one shot collapses the state; re-measuring repeats. Then the Bloch globe.
qm.seed(7)                              # a seed whose first Z-shot reads DOWN, for the story
outcome, collapsed = qm.measure(psi, "Z")
print("this shot reads", "UP (+1)" if outcome == +1 else "DOWN (-1)",
      "->  state snaps to", np.round(collapsed.real, 3))
print("re-measure the same axis:  P =", qm.born_probs(collapsed, "Z")[outcome], " (dead certain)")
qm.plot_bloch([psi, qm.UP, qm.DOWN], labels=["psi (25/75)", "|up>", "|down>"]);
```
