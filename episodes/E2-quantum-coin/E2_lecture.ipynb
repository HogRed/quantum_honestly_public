# The trap: a coin already decided?

> ⚠ **the coin trap (1 of 3)**

The comfortable picture:
- the coin is **already** heads or tails,
- "fifty-fifty" just means *you* haven't peeked.

Comforting. Also **wrong** — and we'll *prove* it, with one twist of a magnet.

# Meet the machine

**The cold-open apparatus, named:**
- **Magnet** = a *sorter* (a **Stern–Gerlach** magnet)
- **Plate** = where the atom lands → top dot = **up**, bottom dot = **down**
- **Dial** = you can *rotate* the magnet

Two dial settings we'll use:
- upright = the **Z setting**
- tipped on its side = the **X setting**

The silver **atom** is our *coin.* The property it sorts is called **spin** — ⚠ *not* a spinning ball (more at the end).

# Why we need a new kind of number

**A line:** ← → only two directions (forward / back = + / −).

**A sheet:** an arrow can point **any** direction.

E1's ± was a number stuck on a line.
To carry *direction* — the thing that made the fringes — we need numbers that live on a **flat sheet.**

# A number you can point — the arrow

An **arrow** has a **length** and a **direction.**

Two ways to pin down the *same* arrow:
1. its **length** + its **direction** (the angle **θ**, "theta")
2. its **east shadow** + its **north shadow**

**length + direction = the same arrow, described two ways.**

*(These point-anywhere numbers have a name: **complex numbers** — just arrows on a sheet.)*

## 🎓 A closer look: cos and sin are just the arrow's shadows

A length-1 arrow at angle θ has an **east-shadow on the floor** and a **north-shadow up the wall.** Those two shadows already have names:

- east-shadow  =  **cos θ**
- north-shadow  =  **sin θ**

That is the whole definition. (For an arrow of length *r*, just scale both: *r*·cos θ and *r*·sin θ.)

**Picture it —** shine a light straight down on the arrow: the puddle of shadow on the floor is *cos*. Shine a light from the side: the shadow it throws up the wall is *sin*. Swing the arrow around and the two shadows simply trade off.

**Watch out:** you do **not** need to "remember trig" to follow this episode. cos and sin are not formulas to memorize — they are just labels for *how far east* and *how far north* the arrow reaches.

<small>Source: unit-circle definitions of cosine and sine (standard trigonometry); cf. Tong, *Quantum Mechanics* §1.1 (complex amplitudes).</small>

## One new tool: arrows that spin

> 🎬 **Animation — i as a quarter-turn:** [▶ watch in EP2 on YouTube](https://youtu.be/RP3VV5buO5c)

Complex numbers are **arrows**. Multiplying by **i** is a **quarter-turn**. The arrow's direction has a name — the **phase**.

> **Leak:** this arrow doesn't spin in *real* space, and its phase is invisible to any single measurement — you only feel it when two arrows interfere.

## 🎓 A closer look: i isn't mystical — it's a rotation

Multiplying by i turns an arrow a **quarter-turn** (90°). So do it **twice** and you've turned a **half-turn** (180°) — you're now facing backwards.

Facing backwards is exactly what multiplying by **−1** does. Therefore:

$$i \cdot i = -1$$

**Picture it —** stand facing east. Quarter-turn left → now facing north. Another quarter-turn left → now facing west, dead opposite where you began. Two left-turns make an about-face, and an about-face is "times minus one."

**Watch out:** "imaginary" is a terrible, scary name for a 400-year-old accident. i is not fake or non-existent — it's just the instruction *"rotate a quarter-turn,"* every bit as real as −1 meaning *"rotate a half-turn."*

<small>Source: the Argand-plane / geometric view of complex numbers (i as a 90° rotation, i² = −1); cf. T. Needham, *Visual Complex Analysis* §1.</small>

## Square the length → the angle drops out

> 🎬 **Animation — squaring the length drops the angle:** [▶ watch in EP2 on YouTube](https://youtu.be/RP3VV5buO5c)

`|z|²` is the **odds**. Two arrows pointing different ways can share the same `|z|²`. *Hold that thought.*

## Amplitudes add, then square (from E1)

> 🎬 **Animation — amplitudes add, then square:** [▶ watch in EP2 on YouTube](https://youtu.be/RP3VV5buO5c)

Aligned → bright (4). Opposite → **dark (0)**. Add the arrows first; *then* square.

## 🎓 A closer look: two kinds of phase

There are really TWO phases hiding under one word. A *global* phase spins the whole state `[a,b]` by the same angle — squaring always kills it, so no experiment can ever see it. A *relative* phase — the lone `i` in the bottom slot of `|+y⟩` — is real physics: it hides from the Z detector but jumps out the moment a detector makes the two arrows ADD before squaring.

**Picture it —** two singers, one for “up,” one for “down.” Delay the whole duet by a second and it sounds identical (global). Slide one singer half a beat against the other and the harmony locks in or clashes (relative).

**Watch out:** “phase is invisible” is only true for the OVERALL spin. The relative phase *between* the two amplitudes is invisible to Z and decisive to X.

<small>Source: Tong §1.1.2 / p7: states differing by a constant phase are equivalent, ψ ≡ e^{iα}ψ, and this "doesn't change the probability distribution P = |ψ|²" (global phase). Web (global vs relative): https://www.bohrium.com/en/sciencepedia/feynman/keyword/global_phase_vs_relative_phase — "global phases are…irrelevant…What really matters is…the difference in phase between different possibilities."</small>

## The state-label: |↑⟩ and |↓⟩

`| … ⟩` is just shorthand for **"the coin is in this state."**

- **|↑⟩**  =  the coin is **definitely up**
- **|↓⟩**  =  the coin is **definitely down**

The whole bracket-shaped package is called a **ket.**

<small>Dirac's pun: the mirror half ⟨ … | is a **bra** — *bra* + *ket* = **bracket**. The bra returns later, when we start asking the coin questions.</small>

## A coin is a two-slot list

Picture a labeled box with two slots — **how much up** on top, **how much down** below:

- **|↑⟩ = [1, 0]** — all up, nothing down
- **|↓⟩ = [0, 1]** — all down, nothing up
- a number out front scales each slot:  **a · [1, 0] = [a, 0]**  drops arrow **a** into the up-slot

This list-of-two has a name: a **vector.**

**⚠ A different "two" than before —** one *arrow* had two *shadows* (east + north — one object). Here each *slot* holds its own *whole arrow.* **Two arrows now — not two shadows.**

## A qubit = two arrows

**|ψ⟩ = a|↑⟩ + b|↓⟩ = a[1,0] + b[0,1] = [a, b]**

- **ψ** ("psi") — the name for the whole state
- **a** = the up-arrow · **b** = the down-arrow
- both at once until you look → **superposition**

**|a|² + |b|² = 1** — the odds add to one (`normalize`)

→ two arrows = the **two numbers** promised

## Two states, written out

**|+x⟩ = (1/√2) [1, 1]**
**|+y⟩ = (1/√2) [1, i]**   ← note the **i** in the down-slot

- `+x`, `+y` = two **sideways** settings, 90° apart (east vs north)
- **same lengths, different directions** (squaring hides the difference)
- **1/√2** = the normalizer (why? → next slide)

**a\*** = the **mirror-image** (complex conjugate) — flip the sign of the `i`

## 🎓 A closer look: where 1/√2 comes from

Want a fair split — equal up, equal down. Start with **[1, 1]**, one in each slot. But square the lengths and add: **1² + 1² = 2**, too big. Shrink both arrows by dividing each by **√2** (the number whose square is 2). Now each square is **(1/√2)² = ½**, and **½ + ½ = 1**. Odds restored.

**Picture it —** cutting one whole pizza into two equal slices: each slice is half the pie. The √2 just resizes the arrows so the two "slices of certainty" total one whole.

**Watch out:** 1/√2 ≈ 0.707 is **not** ½. It's the *square* of the length that's ½ — the arrow's length is 0.707, the odds are 0.5.

<small>Source: David Tong, *Lectures on Quantum Mechanics* (Cambridge), §1 — normalization of states.</small>

## 🎓 A closer look: why |i/√2|² is still a half

The squared length of a complex number is its length squared, full stop — and a length is always real and ≥ 0, so a probability never comes out imaginary or negative. Here `i` is an arrow of length 1 pointed straight up (a pure quarter-turn), so `i/√2` has length `1/√2`; square that length and you get `½`. The `i` only set the arrow's direction — and direction is exactly what squaring throws away.

**Picture it —** spin a pencil flat on a table. Pointing it east, north, or northeast never changes how long the pencil is — and the probability is the pencil's length squared, not where its tip points.

**Watch out:** an imaginary amplitude does NOT give an imaginary or negative probability — probability is `a·a*`, the length squared, always real and ≥ 0.

<small>Source: Tong §3.1.1 / p64: the inner product conjugates the first vector so that ⟨u|u⟩ = u*·u ≥ 0, which lets you "define the length of a vector ||u|| = √⟨u|u⟩" (positive definite). Web (phase rotation preserves the squared norm): https://unstableontology.com/2026/01/09/understanding-complex-conjugates-in-quantum-mechanics/</small>

## Reading the coin — the Born rule

**Overlap** = how much one arrow-pair points along another.

**⟨φ|ψ⟩ = φ\* · ψ** — mirror the first, multiply matching slots, add
- `φ` ("phi") = the outcome you're testing for
- `⟨φ|` = the **bra** — the outcome turned into a question

**P = |⟨φ|ψ⟩|²** — overlap, then square → the odds

<small>the **Born rule** (Max Born) · *why* squared? → E3</small>

## 🎓 A closer look: why conjugate the first arrow

The overlap recipe: line your state `ψ` against an outcome `φ` by multiplying matching slots and adding — but first flip the sign of any `i` in `φ`, the one written first in the bracket. That conjugation isn't decoration: measure a state against ITSELF and the mirror-flip turns the sum into `a·a* + b·b* = |a|² + |b|²`, a real number that equals 1 for a normalized state — certainty, as it must be.

**Picture it —** conjugating-then-multiplying is like taking a distance: `(a+bi)(a−bi) = a² + b²`, always positive because the mirror copy squares away the sign and the angle. It's the complex cousin of `|−3| = 3`.

**Watch out:** you conjugate the FIRST (bra) vector so a state overlapped with itself gives a real, non-negative number — the only kind of number that can be a probability.

<small>Source: Tong §3.1.1 / p64–65: ⟨u|v⟩ = u*·v, conjugate the first "so that the inner product of any vector with itself is necessarily non-negative," and ⟨ψ|φ⟩* = ⟨φ|ψ⟩. Born rule itself: Tong §3.3 / p76, Prob(λ_n) = |a_n|² with ψ = Σ a_n φ_n. Web: https://phys.libretexts.org/Bookshelves/Quantum_Mechanics/Advanced_Quantum_Mechanics_(Kok)/06%3A_Evolution_of_Open_Quantum_Systems/6.03%3A_Bra_Vectors_and_the_Inner_Product</small>

```python
# LIVE: |+x> and |+y> -- identical on Z, different on X
for name, st in [("|+x>", A), ("|+y>", B)]:
    pz, px = qm.born_probs(st, "Z"), qm.born_probs(st, "X")
    print(f"{name}:   Z -> up {pz[+1]:.2f} / down {pz[-1]:.2f}      X -> +x {px[+1]:.2f} / -x {px[-1]:.2f}")
ov = qm.overlap(A, B)
print(f"\noverlap <+x|+y> = {ov:.2f}      |overlap|^2 = {abs(ov)**2:.2f}")
```

## Same odds on Z, different on X

> 🎬 **Animation — same odds on Z, different on X:** [▶ watch in EP2 on YouTube](https://youtu.be/RP3VV5buO5c)

Now let's reproduce it live.

## 🎓 A closer look: why one detector is blind to what another sees

A Z measurement asks “up or down?” — it reads the two arrow-lengths `|a|²` and `|b|²` separately and never combines them, so the angle between the arrows can't matter; both states give a half each way. An X measurement asks a different question, so you must ADD `a` and `b` first and only THEN square. Now the relative phase lives inside the sum: `|+x⟩` adds to overlap 1 (100%), while `|+y⟩`'s quarter-turn makes them add to `½ + i/2`, whose length squared is only `½`.

**Picture it —** two solid objects can throw the SAME shadow on the floor (Z) yet DIFFERENT shadows on the wall (X) — same one way, different the other.

**Watch out:** the qubit is NOT a hidden solid arrow whose shadows you're uncovering — measuring along an axis CREATES the outcome. Matching on Z but splitting on X proves there was no pre-written answer.

<small>Source: Tong §3.3 / p78: measuring a different observable whose eigenstates don't coincide gives a genuinely different, uncertain result — "the quantum particle can't have specific values of Ô and M̂ at the same time." Worked numbers match the script: ⟨+x|+y⟩ = ½ + i/2, |½+i/2|² = ½. Tong §3.1 / p63 thesis: "at heart, quantum mechanics is really a theory of linear algebra."</small>

```python
# LIVE: fire each state through a virtual Stern-Gerlach magnet (10,000 shots)
shots = 10000
panels = [(A, "Z", "|+x>  on Z"), (B, "Z", "|+y>  on Z"),
          (A, "X", "|+x>  on X"), (B, "X", "|+y>  on X")]
fig, axes = plt.subplots(2, 2, figsize=(9, 6))
for ax, (st, axis, title) in zip(axes.ravel(), panels):
    outs = qm.born_sample(st, axis, shots=shots)
    fr = [np.mean(outs == +1), np.mean(outs == -1)]
    ax.bar(["+1", "-1"], fr, color=["#2b8cbe", "#cb4b16"]); ax.set_ylim(0, 1.05); ax.set_title(title)
    for i, f in enumerate(fr): ax.text(i, f + 0.02, f"{f:.2f}", ha="center")
plt.tight_layout(); plt.show()
```

## One honesty note: the half-angle and "spin"

**The half-angle.** On the globe, up and down sit a full **180°** apart — but as arrow-pairs they're only **90°** (a right angle, overlap 0). The globe **doubles** the angle:

math-angle = **θ/2**

**"Spin" is a label, not a spinning ball.** The tell: it takes a **720°** turn — not 360° — to bring a qubit home. Nothing that physically spins does that.

→ Primer: *Spin & the 720° Twist*

## 🎓 A closer look: the half-angle θ/2

Up and down are perfectly tellable apart, which in vector language means orthogonal — a 90° right angle, overlap exactly 0. But on the globe they sit at opposite poles, a full 180° apart. The globe simply DOUBLES every angle (`a = cos(θ/2)`, `b = e^{iφ} sin(θ/2)`): the math-angle `θ/2` between the arrows becomes the angle `θ` you actually walk on the sphere.

**Picture it —** a steering wheel geared 2-to-1 to the tires: crank the wheel a full 180° and the tires turn only 90°. The globe is the wheel; the real quantum state is the tires, turning half as much.

**Watch out:** up and down look 180° apart on the globe, but as state-vectors they're orthogonal — 90° apart, overlap zero. The globe just shows double the true angle.

<small>Source: Tong §1.1.2 / p7: an overall phase and overall scale don't change the physical state (ψ ≡ λψ), which is why two real numbers suffice. Bloch half-angle & antipodal orthogonality: https://qubit.guide/2.10-the-bloch-sphere — |ψ⟩ = cos(θ/2)|0⟩ + sin(θ/2)e^{iφ}|1⟩, "any two orthogonal state vectors appear…as two Bloch vectors pointing in opposite directions." Spin-isn't-a-spinning-ball (720° tell / equator faster than light): https://faraday.physics.utoronto.ca/PVB/Harrison/SternGerlach/SternGerlach.html</small>
