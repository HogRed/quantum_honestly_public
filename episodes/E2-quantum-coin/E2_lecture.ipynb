# Meet the machine

**The cold-open apparatus, named:**
- **Magnet** = a *sorter* (a **Stern‚ÄìGerlach** magnet)
- **Plate** = where the atom lands ‚Üí top dot = **up**, bottom dot = **down**
- **Dial** = you can *rotate* the magnet

Two dial settings we'll use:
- upright = the **Z setting**
- tipped on its side = the **X setting**

The silver **atom** is our *coin.* The property it sorts is called **spin** ‚Äî ‚öÝ *not* a spinning ball (more at the end).

# Why we need a new kind of number

**A line:** ‚Üê ‚Üí only two directions (forward / back = + / ‚àí).

**A sheet:** an arrow can point **any** direction.

E1's ¬± was a number stuck on a line.
To carry *direction* ‚Äî the thing that made the fringes ‚Äî we need numbers that live on a **flat sheet.**

# A number you can point ‚Äî the arrow

An **arrow** has a **length** and a **direction.**

Two ways to pin down the *same* arrow:
1. its **length** + its **direction** (the angle **Œ∏**, "theta")
2. its **east shadow** + its **north shadow**

**length + direction = the same arrow, described two ways.**

*(These point-anywhere numbers have a name: **complex numbers** ‚Äî just arrows on a sheet.)*

## üéì A closer look: cos and sin are just the arrow's shadows

A length-1 arrow at angle Œ∏ has an **east-shadow on the floor** and a **north-shadow up the wall.** Those two shadows already have names:

- east-shadow  =  **cos Œ∏**
- north-shadow  =  **sin Œ∏**

That is the whole definition. (For an arrow of length *r*, just scale both: *r*¬∑cos Œ∏ and *r*¬∑sin Œ∏.)

**Picture it ‚Äî** shine a light straight down on the arrow: the puddle of shadow on the floor is *cos*. Shine a light from the side: the shadow it throws up the wall is *sin*. Swing the arrow around and the two shadows simply trade off.

**Watch out:** you do **not** need to "remember trig" to follow this episode. cos and sin are not formulas to memorize ‚Äî they are just labels for *how far east* and *how far north* the arrow reaches.

<small>Source: unit-circle definitions of cosine and sine (standard trigonometry); cf. Tong, *Quantum Mechanics* ¬ß1.1 (complex amplitudes).</small>

## One new tool: arrows that spin

> üé¨ **Animation ‚Äî i as a quarter-turn:** [‚ñ∂ watch in EP2 on YouTube](https://youtu.be/RP3VV5buO5c)

Complex numbers are **arrows**. Multiplying by **i** is a **quarter-turn**. The arrow's direction has a name ‚Äî the **phase**.

> **Leak:** this arrow doesn't spin in *real* space, and its phase is invisible to any single measurement ‚Äî you only feel it when two arrows interfere.

## üéì A closer look: i isn't mystical ‚Äî it's a rotation

Multiplying by i turns an arrow a **quarter-turn** (90¬∞). So do it **twice** and you've turned a **half-turn** (180¬∞) ‚Äî you're now facing backwards.

Facing backwards is exactly what multiplying by **‚àí1** does. Therefore:

$$i \cdot i = -1$$

**Picture it ‚Äî** stand facing east. Quarter-turn left ‚Üí now facing north. Another quarter-turn left ‚Üí now facing west, dead opposite where you began. Two left-turns make an about-face, and an about-face is "times minus one."

**Watch out:** "imaginary" is a terrible, scary name for a 400-year-old accident. i is not fake or non-existent ‚Äî it's just the instruction *"rotate a quarter-turn,"* every bit as real as ‚àí1 meaning *"rotate a half-turn."*

<small>Source: the Argand-plane / geometric view of complex numbers (i as a 90¬∞ rotation, i¬≤ = ‚àí1); cf. T. Needham, *Visual Complex Analysis* ¬ß1.</small>

## Square the length ‚Üí the angle drops out

`|z|¬≤` is the **odds**. Two arrows pointing different ways can share the same `|z|¬≤`. *Hold that thought.*

## Amplitudes add, then square (from E1)

> üé¨ **Animation ‚Äî amplitudes add, then square:** [‚ñ∂ watch in EP2 on YouTube](https://youtu.be/RP3VV5buO5c)

Aligned ‚Üí bright (4). Opposite ‚Üí **dark (0)**. Add the arrows first; *then* square.

## üéì A closer look: two kinds of phase

There are really TWO phases hiding under one word. A *global* phase spins the whole state `[a,b]` by the same angle ‚Äî squaring always kills it, so no experiment can ever see it. A *relative* phase ‚Äî the lone `i` in the bottom slot of `|+y‚ü©` ‚Äî is real physics: it hides from the Z detector but jumps out the moment a detector makes the two arrows ADD before squaring.

**Picture it ‚Äî** two singers, one for ‚Äúup,‚Äù one for ‚Äúdown.‚Äù Delay the whole duet by a second and it sounds identical (global). Slide one singer half a beat against the other and the harmony locks in or clashes (relative).

**Watch out:** ‚Äúphase is invisible‚Äù is only true for the OVERALL spin. The relative phase *between* the two amplitudes is invisible to Z and decisive to X.

<small>Source: Tong ¬ß1.1.2 / p7: states differing by a constant phase are equivalent, œà ‚â° e^{iŒ±}œà, and this "doesn't change the probability distribution P = |œà|¬≤" (global phase). Web (global vs relative): https://www.bohrium.com/en/sciencepedia/feynman/keyword/global_phase_vs_relative_phase ‚Äî "global phases are‚Ä¶irrelevant‚Ä¶What really matters is‚Ä¶the difference in phase between different possibilities."</small>

## The state-label: |‚Üë‚ü© and |‚Üì‚ü©

`| ‚Ä¶ ‚ü©` is just shorthand for **"the coin is in this state."**

- **|‚Üë‚ü©**  =  the coin is **definitely up**
- **|‚Üì‚ü©**  =  the coin is **definitely down**

The whole bracket-shaped package is called a **ket.**

<small>Dirac's pun: the mirror half ‚ü® ‚Ä¶ | is a **bra** ‚Äî *bra* + *ket* = **bracket**. The bra returns later, when we start asking the coin questions.</small>

## A coin is a two-slot list

Picture a labeled box with two slots ‚Äî **how much up** on top, **how much down** below:

- **|‚Üë‚ü© = [1, 0]** ‚Äî all up, nothing down
- **|‚Üì‚ü© = [0, 1]** ‚Äî all down, nothing up
- a number out front scales each slot:  **a ¬∑ [1, 0] = [a, 0]**  drops arrow **a** into the up-slot

This list-of-two has a name: a **vector.**

**‚öÝ A different "two" than before ‚Äî** one *arrow* had two *shadows* (east + north ‚Äî one object). Here each *slot* holds its own *whole arrow.* **Two arrows now ‚Äî not two shadows.**

## A qubit = two arrows

**|œà‚ü© = a|‚Üë‚ü© + b|‚Üì‚ü© = a[1,0] + b[0,1] = [a, b]**

- **œà** ("psi") ‚Äî the name for the whole state
- **a** = the up-arrow ¬∑ **b** = the down-arrow
- both at once until you look ‚Üí **superposition**

**|a|¬≤ + |b|¬≤ = 1** ‚Äî the odds add to one (`normalize`)

‚Üí two arrows = the **two numbers** promised

## Two states, written out

**|+x‚ü© = (1/‚àö2) [1, 1]**
**|+y‚ü© = (1/‚àö2) [1, i]**   ‚Üê note the **i** in the down-slot

- `+x`, `+y` = two **sideways** settings, 90¬∞ apart (east vs north)
- **same lengths, different directions** (squaring hides the difference)
- **1/‚àö2** = the normalizer (why? ‚Üí next slide)

**a\*** = the **mirror-image** (complex conjugate) ‚Äî flip the sign of the `i`

## üéì A closer look: where 1/‚àö2 comes from

Want a fair split ‚Äî equal up, equal down. Start with **[1, 1]**, one in each slot. But square the lengths and add: **1¬≤ + 1¬≤ = 2**, too big. Shrink both arrows by dividing each by **‚àö2** (the number whose square is 2). Now each square is **(1/‚àö2)¬≤ = ¬Ω**, and **¬Ω + ¬Ω = 1**. Odds restored.

**Picture it ‚Äî** cutting one whole pizza into two equal slices: each slice is half the pie. The ‚àö2 just resizes the arrows so the two "slices of certainty" total one whole.

**Watch out:** 1/‚àö2 ‚âà 0.707 is **not** ¬Ω. It's the *square* of the length that's ¬Ω ‚Äî the arrow's length is 0.707, the odds are 0.5.

<small>Source: David Tong, *Lectures on Quantum Mechanics* (Cambridge), ¬ß1 ‚Äî normalization of states.</small>

## üéì A closer look: why |i/‚àö2|¬≤ is still a half

The squared length of a complex number is its length squared, full stop ‚Äî and a length is always real and ‚â• 0, so a probability never comes out imaginary or negative. Here `i` is an arrow of length 1 pointed straight up (a pure quarter-turn), so `i/‚àö2` has length `1/‚àö2`; square that length and you get `¬Ω`. The `i` only set the arrow's direction ‚Äî and direction is exactly what squaring throws away.

**Picture it ‚Äî** spin a pencil flat on a table. Pointing it east, north, or northeast never changes how long the pencil is ‚Äî and the probability is the pencil's length squared, not where its tip points.

**Watch out:** an imaginary amplitude does NOT give an imaginary or negative probability ‚Äî probability is `a¬∑a*`, the length squared, always real and ‚â• 0.

<small>Source: Tong ¬ß3.1.1 / p64: the inner product conjugates the first vector so that ‚ü®u|u‚ü© = u*¬∑u ‚â• 0, which lets you "define the length of a vector ||u|| = ‚àö‚ü®u|u‚ü©" (positive definite). Web (phase rotation preserves the squared norm): https://unstableontology.com/2026/01/09/understanding-complex-conjugates-in-quantum-mechanics/</small>

## Reading the coin ‚Äî the Born rule

**Overlap** = how much one arrow-pair points along another.

**‚ü®œÜ|œà‚ü© = œÜ\* ¬∑ œà** ‚Äî mirror the first, multiply matching slots, add
- `œÜ` ("phi") = the outcome you're testing for
- `‚ü®œÜ|` = the **bra** ‚Äî the outcome turned into a question

**P = |‚ü®œÜ|œà‚ü©|¬≤** ‚Äî overlap, then square ‚Üí the odds

<small>the **Born rule** (Max Born) ¬∑ *why* squared? ‚Üí E3</small>

## üéì A closer look: why conjugate the first arrow

The overlap recipe: line your state `œà` against an outcome `œÜ` by multiplying matching slots and adding ‚Äî but first flip the sign of any `i` in `œÜ`, the one written first in the bracket. That conjugation isn't decoration: measure a state against ITSELF and the mirror-flip turns the sum into `a¬∑a* + b¬∑b* = |a|¬≤ + |b|¬≤`, a real number that equals 1 for a normalized state ‚Äî certainty, as it must be.

**Picture it ‚Äî** conjugating-then-multiplying is like taking a distance: `(a+bi)(a‚àíbi) = a¬≤ + b¬≤`, always positive because the mirror copy squares away the sign and the angle. It's the complex cousin of `|‚àí3| = 3`.

**Watch out:** you conjugate the FIRST (bra) vector so a state overlapped with itself gives a real, non-negative number ‚Äî the only kind of number that can be a probability.

<small>Source: Tong ¬ß3.1.1 / p64‚Äì65: ‚ü®u|v‚ü© = u*¬∑v, conjugate the first "so that the inner product of any vector with itself is necessarily non-negative," and ‚ü®œà|œÜ‚ü©* = ‚ü®œÜ|œà‚ü©. Born rule itself: Tong ¬ß3.3 / p76, Prob(Œª_n) = |a_n|¬≤ with œà = Œ£ a_n œÜ_n. Web: https://phys.libretexts.org/Bookshelves/Quantum_Mechanics/Advanced_Quantum_Mechanics_(Kok)/06%3A_Evolution_of_Open_Quantum_Systems/6.03%3A_Bra_Vectors_and_the_Inner_Product</small>

```python
# LIVE: |+x> and |+y> -- identical on Z, different on X
for name, st in [("|+x>", A), ("|+y>", B)]:
    pz, px = qm.born_probs(st, "Z"), qm.born_probs(st, "X")
    print(f"{name}:   Z -> up {pz[+1]:.2f} / down {pz[-1]:.2f}      X -> +x {px[+1]:.2f} / -x {px[-1]:.2f}")
ov = qm.overlap(A, B)
print(f"\noverlap <+x|+y> = {ov:.2f}      |overlap|^2 = {abs(ov)**2:.2f}")
```

## Same odds on Z, different on X

> üé¨ **Animation ‚Äî same odds on Z, different on X:** [‚ñ∂ watch in EP2 on YouTube](https://youtu.be/RP3VV5buO5c)

Now let's reproduce it live.

## üéì A closer look: why one detector is blind to what another sees

A Z measurement asks ‚Äúup or down?‚Äù ‚Äî it reads the two arrow-lengths `|a|¬≤` and `|b|¬≤` separately and never combines them, so the angle between the arrows can't matter; both states give a half each way. An X measurement asks a different question, so you must ADD `a` and `b` first and only THEN square. Now the relative phase lives inside the sum: `|+x‚ü©` adds to overlap 1 (100%), while `|+y‚ü©`'s quarter-turn makes them add to `¬Ω + i/2`, whose length squared is only `¬Ω`.

**Picture it ‚Äî** two solid objects can throw the SAME shadow on the floor (Z) yet DIFFERENT shadows on the wall (X) ‚Äî same one way, different the other.

**Watch out:** the qubit is NOT a hidden solid arrow whose shadows you're uncovering ‚Äî measuring along an axis CREATES the outcome. Matching on Z but splitting on X proves there was no pre-written answer.

<small>Source: Tong ¬ß3.3 / p78: measuring a different observable whose eigenstates don't coincide gives a genuinely different, uncertain result ‚Äî "the quantum particle can't have specific values of √î and MÃÇ at the same time." Worked numbers match the script: ‚ü®+x|+y‚ü© = ¬Ω + i/2, |¬Ω+i/2|¬≤ = ¬Ω. Tong ¬ß3.1 / p63 thesis: "at heart, quantum mechanics is really a theory of linear algebra."</small>

```python
# LIVE: fire each state through a virtual Stern-Gerlach magnet (10,000 shots)
shots = 10000
panels = [(A, "Z", "|+x>  on Z"), (B, "Z", "|+y>  on Z"),
          (A, "X", "|+x>  on X"), (B, "X", "|+y>  on X")]
fig, axes = plt.subplots(2, 2, figsize=(9, 6))
for ax, (st, axis, title) in zip(axes.ravel(), panels):
    outs = qm.born_sample(st, axis, shots=shots)
    fr = [np.mean(outs == +1), np.mean(outs == -1)]
    ax.bar(["+1", "-1"], fr, color=["#2b8cbe", "#cb4b16"]); ax.set_ylim(0, 1.05); ax.set_title(title)
    for i, f in enumerate(fr): ax.text(i, f + 0.02, f"{f:.2f}", ha="center")
plt.tight_layout(); plt.show()
```

## One honesty note: the half-angle and "spin"

**The half-angle.** On the globe, up and down sit a full **180¬∞** apart ‚Äî but as arrow-pairs they're only **90¬∞** (a right angle, overlap 0). The globe **doubles** the angle:

math-angle = **Œ∏/2**

**"Spin" is a label, not a spinning ball.** The tell: it takes a **720¬∞** turn ‚Äî not 360¬∞ ‚Äî to bring a qubit home. Nothing that physically spins does that.

‚Üí Primer: *Spin & the 720¬∞ Twist*

## üéì A closer look: the half-angle Œ∏/2

Up and down are perfectly tellable apart, which in vector language means orthogonal ‚Äî a 90¬∞ right angle, overlap exactly 0. But on the globe they sit at opposite poles, a full 180¬∞ apart. The globe simply DOUBLES every angle (`a = cos(Œ∏/2)`, `b = e^{iœÜ} sin(Œ∏/2)`): the math-angle `Œ∏/2` between the arrows becomes the angle `Œ∏` you actually walk on the sphere.

**Picture it ‚Äî** a steering wheel geared 2-to-1 to the tires: crank the wheel a full 180¬∞ and the tires turn only 90¬∞. The globe is the wheel; the real quantum state is the tires, turning half as much.

**Watch out:** up and down look 180¬∞ apart on the globe, but as state-vectors they're orthogonal ‚Äî 90¬∞ apart, overlap zero. The globe just shows double the true angle.

<small>Source: Tong ¬ß1.1.2 / p7: an overall phase and overall scale don't change the physical state (œà ‚â° Œªœà), which is why two real numbers suffice. Bloch half-angle & antipodal orthogonality: https://qubit.guide/2.10-the-bloch-sphere ‚Äî |œà‚ü© = cos(Œ∏/2)|0‚ü© + sin(Œ∏/2)e^{iœÜ}|1‚ü©, "any two orthogonal state vectors appear‚Ä¶as two Bloch vectors pointing in opposite directions." Spin-isn't-a-spinning-ball (720¬∞ tell / equator faster than light): https://faraday.physics.utoronto.ca/PVB/Harrison/SternGerlach/SternGerlach.html</small>
